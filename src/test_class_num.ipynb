# check sample N per class 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../input/bengaliai-cv19/train.csv')
train_labels = train[['grapheme_root', 'vowel_diacritic', 'consonant_diacritic']].values
train_labels.shape
```

```python
COUNT_grapheme_root = \
train.groupby(by=['grapheme_root']).count().reset_index()[['grapheme_root', 'image_id']]

COUNT_vowel_diacritic = \
train.groupby(by=['vowel_diacritic']).count().reset_index()[['vowel_diacritic', 'image_id']]

COUNT_consonant_diacritic = \
train.groupby(by=['consonant_diacritic']).count().reset_index()[['consonant_diacritic', 'image_id']]

# print to hard-code later
print(COUNT_grapheme_root.image_id.values.tolist())
print(COUNT_vowel_diacritic.image_id.values.tolist())
print(COUNT_consonant_diacritic.image_id.values.tolist())
```

```python
plt.figure(figsize=(20, 15))
plt.subplot(3,1,1), plt.grid(True)
plt.bar(range(0, len(COUNT_grapheme_root)), COUNT_grapheme_root.image_id.values)
plt.subplot(3,1,2), plt.grid(True)
plt.bar(range(0, len(COUNT_vowel_diacritic)), COUNT_vowel_diacritic.image_id.values)
plt.subplot(3,1,3), plt.grid(True)
plt.bar(range(0, len(COUNT_consonant_diacritic)), COUNT_consonant_diacritic.image_id.values)
```
